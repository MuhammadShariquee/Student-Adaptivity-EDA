```python
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("students_adaptability_.csv")
```

```python
df
```

```python
df.isnull().sum()
```

```python
df.info()
```

```python
df["Education Level"] = df["Education Level"].map({'School': 0, 'College': 1, 'University': 2})
```

```python
df.head()
```

```python
df["Adaptivity Level"].unique()
```

```python
df["Adaptivity Level"] = df["Adaptivity Level"].map({'Low': 0, 'Moderate': 1, 'High': 2})
```

```python
df["Institution Type"] = df["Institution Type"].map({'Government': 0, 'Non Government': 1})
```

```python
df
```

```python
sns.countplot(df, x = 'Adaptivity Level') 
# task: 1.Distribution of Adaptivity Level
```

```python
sns.countplot(df, x='Gender', hue = 'Age') 
#task: 1.Distribution of Gender
```

```python

sns.histplot(df, x = 'Age', hue='Gender')
```

```python

sns.histplot(data=df, x='Age', hue='Financial Condition', multiple='stack', bins=10, palette='coolwarm')
plt.title('Age vs Financial Condition')
plt.xlabel('Age')
plt.ylabel('Count')
plt.show()
```

```python
sns.scatterplot(data=df, x='Education Level', y='Adaptivity Level', alpha=0.7, s=60)

plt.title('Relationship Between Education Level and Adaptivity Level')
plt.xlabel('Education Level (0=School, 1=College, 2=University)')
plt.ylabel('Adaptivity Level (0=Low, 1=Moderate, 2=High)')
plt.show()
```

```python
sns.barplot(data=df, x='Education Level', y='Adaptivity Level')
```

```python
sns.barplot(df, x ='Institution Type', y='Adaptivity Level')
```

```python
df
```

```python
sns.lineplot(df, x= 'Education Level', y = 'Adaptivity Level', hue = 'Gender')
plt.grid(True)
```

```python
sns.boxplot(data=df, x='Gender', y='Age')
```

```python
sns.pairplot(df)
```

```python
df['Gender'].value_counts().plot(kind='pie', autopct='%1.1f%%')
plt.title('Gender Distribution')
plt.show()
```

<h1>Conclusion</h1>

<p>This exploratory data analysis provided useful insights into how different student factors relate to their adaptability levels. We observed clear patterns across education level, institution type, gender, age, and financial conditions. The visualizations helped highlight trends such as how adaptability varies with education level and demographics while also revealing the distribution of key features in the dataset. Overall, the EDA gives a strong foundation for understanding the dataset and can be used for future predictive or analytical work.</p>

<h2>Key Insights From the Analysis</h2>

<p><b>University students</b> show the highest adaptivity levels, indicating they adjust better to online learning than school or college students.

<b>Higher education level</b> is positively related to adaptivity  as education level increases, adaptivity also tends to increase.

<b>Institution type</b> has a noticeable impact  students from government institutions show different adaptivity trends compared to non-government ones.

<b>Age distribution</b> is varied, but no extreme imbalance is found; most students belong to a middle age range.

<b>Gender distribution</b> appears balanced, with no major difference in basic counts.

<b>Financial condition</b> affects age patterns, with diverse representation across age groups.

Adaptivity level overall is fairly distributed, but moderate and high adaptivity appear more frequent</p>
