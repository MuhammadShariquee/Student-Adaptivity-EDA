```python
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("students_adaptability_.csv")
```

```python
df
```

```python
df.isnull().sum()
```

```python
df.info()
```

```python
df["Education Level"] = df["Education Level"].map({'School': 0, 'College': 1, 'University': 2})
```

```python
df.head()
```

```python
df["Adaptivity Level"].unique()
```

```python
df["Adaptivity Level"] = df["Adaptivity Level"].map({'Low': 0, 'Moderate': 1, 'High': 2})
```

```python
df["Institution Type"] = df["Institution Type"].map({'Government': 0, 'Non Government': 1})
```

```python
df
```

```python
sns.countplot(df, x = 'Adaptivity Level') 
# task: 1.Distribution of Adaptivity Level
```

```python
sns.countplot(df, x = 'Gender', hue = 'Age') 
# task: 1.Distribution of Gender
```

```python

sns.histplot(df, x = 'Age', hue='Gender')
```

```python

sns.histplot(data=df, x='Age', hue='Financial Condition', multiple='stack', bins=10, palette='coolwarm')
plt.title('Age vs Financial Condition')
plt.xlabel('Age')
plt.ylabel('Count')
plt.show()
```

```python
sns.scatterplot(data=df, x='Education Level', y='Adaptivity Level', alpha=0.7, s=60)

plt.title('Relationship Between Education Level and Adaptivity Level')
plt.xlabel('Education Level (0=School, 1=College, 2=University)')
plt.ylabel('Adaptivity Level (0=Low, 1=Moderate, 2=High)')
plt.show()
```

```python
sns.barplot(data=df, x='Education Level', y='Adaptivity Level')
```

```python
sns.barplot(df, x ='Institution Type', y='Adaptivity Level')
```
